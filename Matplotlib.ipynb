# Matplotlib'e Giriş

```python
import numpy as np # numpy import
import pandas as pd # pandas import
import matplotlib.pyplot as plt # matplot import
```

```python
# Plotting two 1-D numpy arrays
x = np.linspace(5, 100, 100) # 5 ve 100 arası 100 veri
y = np.linspace(10, 1000, 100) # 10 ile 1000 arası 100 veri

plt.plot(x, y) # ve plot
```

```python
# plotu göstermek.
plt.show()
```

```python
# ayrıca list vererek de plot çizilebilir.
plt.plot([1, 4, 6, 8], [3, 8, 3, 5])
plt.show()
```

```python
# x ve y ekseni oluşturma
plt.plot(x, y)

# x ve y ekseninin etiketlerinin yazılması
plt.xlabel("Current")
plt.ylabel("Voltage")
plt.title("Ohm's Law") # plot başlığı

# x ve y eksenlerini limitlemek.
plt.xlim([20, 80])
plt.ylim([200, 800])
plt.show()
```

```python
# renk ve çizgi tipini değiştirmek.

# x ve y dizilerini oluşturmak.
x = np.linspace(0, 10, 20)
y = x*2

# çizgi rengini mavi ve tipini artı işaretine çevirelim.
plt.plot(x, y, 'b+')

# tekrar x ve y eksenine etiketlerini ekleyelim.
plt.xlabel("Current")
plt.ylabel("Voltage")
plt.title("Ohm's Law")

plt.show()
```

```python
# birden fazla çizgiyi aynı plot'a koymak.
x = np.linspace(0, 5, 10) # dizi oluşturma
y = np.linspace(3, 6, 10) # dizi oluşturma.

# 3 farklı eğriyi çizmek: y, y**2 and y**3 
plt.plot(x, y, 'r-', x, y**2, 'b+', x, y**3, 'g^')
plt.show()
```

```python
plt.subplot(1,2,1) #1 satır,2 sütunlu grafikten 1.grafik demek
plt.plot(x,y,'r')
plt.subplot(1,2,2) #1 satır,2 sütunlu grafikten 2.grafik demek
plt.plot(x,z,'y')
# Yukarıda çizdiğimiz grafikleri iç içe girmeden düzgün şekilde çizdirmek
#istersek
plt.tight_layout()
```

```python
x = np.linspace(1, 10, 100) # array 1
y = np.log(x) # array 2

# figürü initialize etme.
plt.figure(1)

# 1 satır ve 2 kolonlu plot oluşturmak.

# ilk plot'un figürde çizdirilmesi.
plt.subplot(121)     #  plt.subplot(1, 2, 1)'a eşittir. 
plt.title("y = log(x)")
plt.plot(x, y)

# ikinci plot'u figür 1'de çizdirmek.
plt.subplot(122)
plt.title("y = log(x)**2")
plt.plot(x, y**2)
plt.show()
```

```python
# örnek data'nın oluşturulması 100 adet 1 ve 10 arasında.
x = np.linspace(1, 10, 100)

plt.figure(1)

# subplot 1
plt.subplot(2, 2, 1)
plt.title("Linear")
plt.plot(x, x)


# subplot 2 
plt.subplot(2, 2, 2)
plt.title("Cubic")
plt.plot(x, x**3)


# subplot 3
plt.figure(2)

plt.subplot(2, 2, 1)
plt.title("Log")
plt.plot(x, np.log(x))

# subplot 4 
plt.subplot(2, 2, 2)
plt.title("Exponential")
plt.plot(x, x**2)


plt.show()

```

# Genel olarak kullanılan chart tipleri

```python
# Sales datasının içeriye alınması
df = pd.read_csv("./global_sales_data/market_fact.csv")
df.head()
```

# Scatter
İki değişkenin ilişkisini göstermeye yarayan plottur.

```python
# Sales ve profit değişkenlerinin görselleştirilmesi.
plt.scatter(df['Sales'], df['Profit'])
plt.show()
```

# Histogram
Bir değişkenin dağılımını görmek için kullanılır.

```python
# Histograms

plt.hist(df['Sales'])
plt.show()
```

```python
# Histogramı y scale log yaparsak daha okunabilir olur.

plt.hist(df['Sales'])
plt.yscale('log')
plt.show()
```

# Boxplot

```python
# Sürekli değişken Order Quantity'nin boxplot gösterimi.
plt.boxplot(df['Order_Quantity'])
plt.show()
```

```python
# Sales için boxplot okunbilir değildir.
plt.boxplot(df['Sales'])
plt.show()
```

Bunun nedeni verinin çok büyük ve küçük değerler içermesidir.

```python
# Eğer verinin istatistiklerine bakarsak maximum 89 bin minimum 2'dir.
df['Sales'].describe()
```

Bunu çözmek için logaritmik fonksiyon kullanarak ölçüleme yapabiliriz. 
plt.yscale('log').

```python
# Orjinal Sales
plt.subplot(1, 2, 1)
plt.boxplot(df['Sales'])


# Log(Sales) box plot
plt.subplot(1, 2, 2)
plt.boxplot(df['Sales'])
plt.yscale('log')
plt.show()
```

# PieChart

```python
labels = 'Türkiye','ABD','Çin','Almanya'
sizes = [15,35,35,15]
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

# explode kısmındaki sayılar grafikteki payların ne kadar 
#uzak görünebileceğine karar verir.
# explode = (0, 0, 0, 0) bu şekilde olduğunda hepsi beraber 
# yapışık görünür.

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
